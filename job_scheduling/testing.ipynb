```python
from graph import Graph
from ant_colony_optimizer import AntColonyOptimizer

import pandas as pd
import numpy as np
from functools import partial
```

## Small graph

```python
job_labels_1 = ['A', 'B', 'C', 'D', 'E']
job_deadlines_1 = [2, 1, 2, 1, 3]
job_profits_1 = [100, 25, 27, 19, 15]

job_1 = pd.DataFrame({'Labels': job_labels_1,
                      'Deadlines': job_deadlines_1,
                      'Profits': job_profits_1})

graph_1 = Graph()
graph_1.generate_from_job(job_1)
aco_1 = AntColonyOptimizer(graph_1, ants=5, evaporation_rate=0.3, intensification=0.5, tmax=100, alpha=0.3, beta=1)
aco_1.fit()
```
